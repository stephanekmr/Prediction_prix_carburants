Calcul des prix moyens par année pour chaque type de carburant

```python
# Fonction pour calculer les prix moyens pour chaque type de carburant
def prix_moyen_par_annee (type_carburant):
     dico = {
        '2007' : df_2007[df_2007['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2008' : df_2008[df_2008['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2009' : df_2009[df_2009['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2010' : df_2010[df_2010['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2011' : df_2011[df_2011['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2012' : df_2012[df_2012['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2013' : df_2013[df_2013['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2014' : df_2014[df_2014['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2015' : df_2015[df_2015['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2016' : df_2016[df_2016['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2017' : df_2017[df_2017['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2018' : df_2018[df_2018['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2019' : df_2019[df_2019['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2020' : df_2020[df_2020['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2021' : df_2021[df_2021['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2022' : df_2022[df_2022['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2023' : df_2023[df_2023['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2024' : df_2024[df_2024['type_carburant'] == type_carburant]['prix'].mean()/1000,
        '2025' : df_2025[df_2025['type_carburant'] == type_carburant]['prix'].mean()/1000,
        # 'flux' : df_flux[df_flux['type_carburant'] == type_carburant]['prix'].mean()/1000
     }
     return dico
```

```python
# On crée les différents dictionnaires de prix moyens pour chaque type de carburant
dico_prix_moyen_Gazole = prix_moyen_par_annee('Gazole')
dico_prix_moyen_SP95 = prix_moyen_par_annee('SP95')
dico_prix_moyen_GPLc = prix_moyen_par_annee('GPLc')
dico_prix_moyen_E85 = prix_moyen_par_annee('E85')
dico_prix_moyen_SP98 = prix_moyen_par_annee('SP98')
dico_prix_moyen_E10 = prix_moyen_par_annee('E10')
```

```python
# Méthode pour tracer le graphe
def tracer_graphe_prix_moyen_par_annee ():
    plt.figure(figsize=(12, 8))
    plt.plot(dico_prix_moyen_Gazole.keys(), dico_prix_moyen_Gazole.values(),marker='o', label='Gazole', color='blue')
    plt.plot(dico_prix_moyen_SP95.keys(), dico_prix_moyen_SP95.values(),marker='o', label='SP95', color='orange')
    plt.plot(dico_prix_moyen_GPLc.keys(), dico_prix_moyen_GPLc.values(),marker='o', label='GPLc', color='green')
    plt.plot(dico_prix_moyen_E85.keys(), dico_prix_moyen_E85.values(),marker='o', label='E85', color='red')
    plt.plot(dico_prix_moyen_SP98.keys(), dico_prix_moyen_SP98.values(),marker='o', label='SP98', color='purple')
    plt.plot(dico_prix_moyen_E10.keys(), dico_prix_moyen_E10.values(),marker='o', label='E10', color='brown')
    plt.title("Évolution des prix moyens des carburants par année (2007 - 2025)")
    plt.xlabel("Année")
    plt.ylabel("Prix moyen (€ par litre)") 
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()
```

Calcul des prix moyens par ville et par année pour chaque type de carburant

```python
# Méthode pour calculer les prix moyens pour chaque type de carburant par ville et par année
def calculer_prix_par_ville_par_annee(dataframe):
    # Regrouper par 'ville' et 'type_carburant', puis calculer la moyenne des prix
    donnees_groupees = dataframe.groupby(['ville', 'type_carburant'])['prix'].mean().reset_index()
    # Diviser les prix par 1000 (pour les avoir en €)
    donnees_groupees['prix'] = donnees_groupees['prix'] / 1000
    # Réorganiser les colonnes
    result =  donnees_groupees.pivot(index='ville', columns='type_carburant', values='prix').reset_index()
    return result.rename_axis(None, axis=1)  # Enlèver l'index de la colonne
```

```python
# On calcule les prix moyens pour chaque type de carburant par ville et par année
df_prix_par_ville_2007 =  calculer_prix_par_ville_par_annee(df_2007)
df_prix_par_ville_2008 =  calculer_prix_par_ville_par_annee(df_2008)
df_prix_par_ville_2009 =  calculer_prix_par_ville_par_annee(df_2009)
df_prix_par_ville_2010 =  calculer_prix_par_ville_par_annee(df_2010)
df_prix_par_ville_2011 =  calculer_prix_par_ville_par_annee(df_2011)
df_prix_par_ville_2012 =  calculer_prix_par_ville_par_annee(df_2012)
df_prix_par_ville_2013 =  calculer_prix_par_ville_par_annee(df_2013)
df_prix_par_ville_2014 =  calculer_prix_par_ville_par_annee(df_2014)
df_prix_par_ville_2015 =  calculer_prix_par_ville_par_annee(df_2015)
df_prix_par_ville_2016 =  calculer_prix_par_ville_par_annee(df_2016)
df_prix_par_ville_2017 =  calculer_prix_par_ville_par_annee(df_2017)
df_prix_par_ville_2018 =  calculer_prix_par_ville_par_annee(df_2018)
df_prix_par_ville_2019 =  calculer_prix_par_ville_par_annee(df_2019)
df_prix_par_ville_2020 =  calculer_prix_par_ville_par_annee(df_2020)
df_prix_par_ville_2021 =  calculer_prix_par_ville_par_annee(df_2021)
df_prix_par_ville_2022 =  calculer_prix_par_ville_par_annee(df_2022)
df_prix_par_ville_2023 =  calculer_prix_par_ville_par_annee(df_2023)
df_prix_par_ville_2024 =  calculer_prix_par_ville_par_annee(df_2024)
df_prix_par_ville_2025 =  calculer_prix_par_ville_par_annee(df_2025)
```

Identifier pour chaque type de carburant les 3 villes où les prix sont les plus bas et celles où les prix étaient les plus élevés

```python
# Méthode pour récupérer les 3 villes où le prix tait le plus bas et celles où le prix était le plus élevé
def calculer_3moins_chers_et_3plus_chers(typeCarburant, dataframe):
    df = dataframe[['ville', typeCarburant]]
    # On supprime les NaN
    df = df.dropna()
    df = df.sort_values(typeCarburant, ascending=True)
    result = pd.concat([df.head(3), df.tail(3)], axis=0)
    return result
```

```python
# Méthode pour afficher les résultats sous forme de bar chart horizotale*
def tracer_graphique_3moins_chers_et_3plus_chers(df_annee, type_caburant, annee) :
    df = calculer_3moins_chers_et_3plus_chers(type_caburant, df_annee)
    plt.figure(figsize=(10, 6))
    plt.barh(df['ville'], df[type_caburant], color=['lightgreen', 'limegreen', 'darkgreen', 'lightcoral', 'red', 'darkred'])
    plt.title(f"Prix du {type_caburant} par ville en {annee} (3 plus chères et 3 moins chères)")
    plt.xlabel("prix (€ par litre)")
    plt.ylabel("Ville")
    plt.xticks (np.arange(0, df[type_caburant].max() + 0.1, 0.1))
    plt.tight_layout()
    plt.show()
```

```python
# On affiche les résultat pour le Gazole en 2007
tracer_graphique_3moins_chers_et_3plus_chers(df_prix_par_ville_2007, 'Gazole', 2007)
```

```python
# On affiche les résultat pour le Gazole en 2024
tracer_graphique_3moins_chers_et_3plus_chers(df_prix_par_ville_2024, "Gazole", 2024)
```

```python
# On affiche les résultat pour le Gazole en 2025
tracer_graphique_3moins_chers_et_3plus_chers(df_prix_par_ville_2025, "Gazole", 2025)
```
